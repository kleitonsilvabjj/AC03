```python
from itertools import permutations
import matplotlib.pyplot as plt

def vector2Distance(a, b):
  return ((a[0] - b[0])**2 + (a[1] - b[1])**2)**0.5

def caixeiroViajante(coords):

  coords = permutations(coords)

  coordenadasSomas = {}
  possiveisSomas = []
  for p in coords:
    somaTotal = 0

    for i in range(len(p) - 1):
      somaTotal += vector2Distance(p[i], p[i+1])

    possiveisSomas.append(somaTotal)
    coordenadasSomas.update({ somaTotal : p })

  menor = coordenadasSomas.get(min(possiveisSomas))

  return menor,possiveisSomas
  
def plotar(caminho,tamanho,label):
  x = []
  y = []

  for i in range(len(caminho)):
    x.append(caminho[i][0])
    y.append(caminho[i][1])
  
  ax = plt.plot(x, y, label=label)
  plt.xlim(0, tamanho)
  plt.ylim(0, tamanho)
  plt.legend()
  plt.show()
```

```python
x = caixeiroViajante([[5,1],[3,2],[4,3],[2,3]])
plotar(x[0],7,"4 cidades")
```

```python
x = caixeiroViajante([[6,2],[3,1],[2,1],[4,5],[3,5]])
plotar(x[0],8,"5 cidades")
```
